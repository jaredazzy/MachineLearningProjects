```python
# Import dependencies
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Load Dataset
car_data = pd.read_csv("car_data.csv")

```

```python
X = car_data.dropna().drop(columns=["Make", "Model", "Market Category", "Engine Fuel Type", "Driven_Wheels", "Transmission Type", "Vehicle Style", "MSRP"], axis = 1)
# SVM model only deals with numbers so any non-number values like vehicle size have to be replaced
X['Vehicle Size'].replace(to_replace={"Compact":0, "Midsize":1, "Large":2}, inplace = True)
X
```

```python
Y = []
budget = 20000 # how much you are willing to spend on a car

car_prices = car_data.dropna()["MSRP"] # extract the price column from our data

for i in range(50):
  if type (car_prices[i] is not int):
    next
  if (car_prices[i] > budget):
    Y.append(0) # not in budget
  elif (car_prices[i] <= budget):
    Y.append(1)

```

```python
# We split the data into training and testing data
# training data will create model and testing data will tell us how good our training was
X_train, X_test, Y_train, Y_test = train_test_split(X[:50], Y, test_size = 0.2, random_state = 2)
print(X_train)
```
